```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')
x = [1,5,4]
y = [21,25,14]
x2 = [4,6,41]
y2 = [5,4,3]

plt.plot(x,y, label="first", linewidth=4)
plt.plot(x2,y2, label="second")
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')
x = [41,58,44]
y = [21,25,14]
x2 = [4,6,41]
y2 = [5,4,3]

plt.hist(x, label="first", linewidth=4)
plt.scatter(x2,y2, label="second")
plt.legend()
plt.show()
```
